```python
import matplotlib.pyplot as plt
import numpy as np

# Nombres de los vrtices
vertices = [
    "Diseo", "Educacin", "Principios ticos",
    "Regulacin", "Manejo de Riesgos", "Diseo"
]

# Calificaciones
calificaciones_wire_shark = [4, 5, 3, 3, 4, 4]
calificaciones_vera_ai = [5, 5, 5, 4.5, 4, 5]

# Crear ngulos para el hexgono
angles = np.linspace(0, 2 * np.pi, len(vertices))

# Crear la figura y los ejes polares
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={"projection": "polar"})

# Graficar el hexgono para cada conjunto de calificaciones
ax.plot(angles, calificaciones_wire_shark, label="WIRE-SHARK", marker="o", color="blue")
ax.plot(angles, calificaciones_vera_ai, label="VERA AI", marker="o", color="green")

# Rellenar el rea entre las calificaciones
ax.fill(angles, calificaciones_wire_shark, alpha=0.2, color="blue")
ax.fill(angles, calificaciones_vera_ai, alpha=0.2, color="green")

# Configurar los nombres de los vrtices
ax.set_xticks(angles)
ax.set_xticklabels(vertices, fontsize=10)

# Configurar lmites del radio
ax.set_ylim(0, 5.5)

# Aadir etiquetas generales para los ejes
ax.text(1, 5.5, "WIRE-SHARK", fontsize=14, color="blue", ha="center", va="center")  # Derecha
ax.text(np.pi, 5.5, "VERA AI", fontsize=14, color="green", ha="center", va="center")  # Izquierda

# Aadir leyenda
ax.legend(loc="upper right", bbox_to_anchor=(1.1, 1.1))

# Ttulo del grfico
ax.set_title("Radar IA", fontsize=14, pad=20)

# Mostrar grfico
plt.show()
```
